## Algorithm

```
# We will pass the food type which can be meal or snack and also the total calories 
# that have been consumed in the day
def get_food(food_type, cals, max_cals):
    # Let's check if the person has consumed all of their calories for the day
    if (cals >= max_cals):
        return "You have consumed all your calories for the day!"

    # Check the type of food
    if (food_type == 'meal'):
        food = meals
        desc = meals_desc
        perc = 0.26
    elif (food_type == 'snack'):
        food = snacks
        desc = snacks_desc
        perc = 0.1
    else:
        return "Please enter snacks or meal"

    # Assume each snack is 10% of max cals so each meal is aprox. 26%
    cals_to_consume = perc * max_cals
    # Get the food to consume
    food_options_idx = food['calories'].sub(cals_to_consume).abs().idxmin()
    return desc.loc[food_options_idx]
```
